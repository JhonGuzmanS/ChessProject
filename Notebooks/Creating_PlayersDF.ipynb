```python
df = pd.read_csv("upset_dataset.csv")
df
```

```python
df.sort_values(by=["elo_diff"])
```

```python
white_list = df["white_id"]
black_list = df["black_id"]
```

```python
user_list = set(pd.concat([white_list, black_list]))
user_list
```

```python
user_info = {}
index = -1
for user in user_list:
    index = index + 1
    matches = df[df["white_id"] == user]
    wins = matches[matches["winner"] == "white"]
    losses = matches[matches["winner"] == "black"]
    user_info[index] = set(matches["white_id"]), list(matches["white_rating"]), set(matches["white_category"]), set(matches["opening_name"]), matches.value_counts().sum(), list(matches["index"]), wins.value_counts().sum(), list(wins["index"])
user_info
```

```python
user_info2 = {}
index = -1
for user in user_list:
    index = index + 1
    matches = df[df["white_id"] == user]
    user_info2[index] = set(matches["black_id"]), list(matches["black_rating"]), set(matches["black_category"]), set(matches["opening_name"]), matches.value_counts().sum(), list(matches["index"]), wins.value_counts().sum(), list(wins["index"])
user_info2
```

```python
cols = ['player_id', 'player_rating', 'category', 'openings played', 'num_matches', 'match_index', 'num_wins', 'win_index']
black_users = pd.DataFrame.from_dict(user_info2, orient='index', columns=cols)
white_users = pd.DataFrame.from_dict(user_info, orient='index', columns=cols)
black_users
```

```python
for i in range(len(white_users)):
    if len(white_users.player_id[i]) == 0:
        white_users.player_id[i] = np.nan

white_users.dropna()
```

```python
for i in range(len(black_users)):
    if len(black_users.player_id[i]) == 0:
        black_users.player_id[i] = np.nan

black_users.dropna()
```

```python
black_users = black_users.dropna()
white_users = white_users.dropna()
```

```python
def remove_list(x):
    return "N/A" if not x else x

white_users["win_index"] = white_users["win_index"].apply(remove_list)
black_users["win_index"] = black_users["win_index"].apply(remove_list)
white_users
```

```python
def convert_str(x):
    return str(x).replace('{', "").replace('}', "").replace("'", "")

white_users["player_id"] = white_users["player_id"].apply(convert_str)
black_users["player_id"] = black_users["player_id"].apply(convert_str)
white_users
```
