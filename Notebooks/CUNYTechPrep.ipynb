```python
#A difference in 200 would in level and the lower level winning would be count as an upset
df_filter=df[(df["elo_diff"]>200) |( df["elo_diff"]>200)]
df_filter.reset_index(inplace=True)
df_upsets=pd.DataFrame()
keep=False
keep_List=[]

df_upset=pd.DataFrame()
for i in range(len(df_filter)):
    if( df_filter.loc[i,"white_rating"]<df_filter.loc[i,"black_rating"] ) & (df_filter.loc[i,"winner"]=="white"):
        keep=True
    if( df_filter.loc[i,"white_rating"]<df_filter.loc[i,"black_rating"] ) & (df_filter.loc[i,"winner"]=="black"):
        keep=False
    if( df_filter.loc[i,"black_rating"]<df_filter.loc[i,"white_rating"] ) & (df_filter.loc[i,"winner"]=="black"):
        keep=True
    if( df_filter.loc[i,"black_rating"]<df_filter.loc[i,"white_rating"] ) & (df_filter.loc[i,"winner"]=="white"):
        keep=False
    keep_List.append(keep)
    keep=False
    
        
```

```python
df_upsets=df_filter[keep_List]
```

```python
df_upsets.reset_index()
```

```python

match_history = df_upsets['moves']
match_history = match_history.dropna()
match_history=match_history.reset_index()
match_history
```
